### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: Window is a figure in matplotlib. Ex: we use (fig = plt.figure() or fig = plt.subplots() ) to draw a figure

Axes:  axes is a elements of plot. Example: ax.plot([1,2,3],[4,5,6])

Axis: axis typically refers to either the x or y dimension of an object, and a plot can contain several axis objects. For example:

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


Artist: An Artist represents any drawable object within a figure. For instance, the line created below is an artist:

line = ax.plot([1, 2, 3], [4, 5, 6])  # The line is an artist

```python
fig, ax = plt.subplots()             
ax.plot([2, 3, 4], [1, 4, 2])
plt.show()
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y = x**2
plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")

plt.title("figure Y =x^2")


```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2


fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python


fig, (ax_A, ax_b) =plt.subplots(1,2, figsize =(13,6))

ax_A.scatter(random_data_x, random_data_y, color ="red")
ax_A.set_title=("Figure: Scatter")
ax_A.set_xlabel=("x")
ax_A.set_ylabel=("y")

ax_b.bar(names, values, color ="yellow")
ax_b.set_title=("Figure:bar")
ax_b.set_xlabel=("name")
ax_b.set_ylabel=("value")






```
